```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as sns
import seaborn as sns
```

```python
#Load the scaled dataset from part 2
url='https://raw.githubusercontent.com/kuNalsharma0725/Capstone-Project--AI_ML/refs/heads/main/scaledd_data.csv'
df=pd.read_csv(url)
print(df.shape)
df.head()
```

```python
#split fratures and targets
y_clf=df['extra_curricular_activity']
X=df.drop('extra_curricular_activity',axis=1)
print(X.shape)
print(y_clf.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train_scaled, X_test_scaled, y_clf_train, y_clf_test=train_test_split(X,y_clf,test_size=0.2,random_state=42)
print(X_train_scaled.shape)
print(X_test_scaled.shape)
print(y_clf_train.shape)
print(y_clf_test.shape)
```

```python
#1.Decision Tree baseline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report, f1_score,precision_score,recall_score
```

```python
#Train the model
dec_tree_model=DecisionTreeClassifier(random_state=42)
dec_tree_model.fit(X_train_scaled,y_clf_train)
```

```python
#Predict the values on both train and test
y_dec_tree_train_pred=dec_tree_model.predict(X_train_scaled)
y_dec_tree_test_pred=dec_tree_model.predict(X_test_scaled)
```

```python
#Report training accuracy and test accuracy
print('Training accuracy:',accuracy_score(y_clf_train,y_dec_tree_train_pred))
print('Test accuracy:',accuracy_score(y_clf_test,y_dec_tree_test_pred))
```

```python
# 2. Controlled Decision Tree
con_dec_tree_model=DecisionTreeClassifier(max_depth=5, min_samples_split=10,random_state=42)
con_dec_tree_model.fit(X_train_scaled,y_clf_train)
```

```python
#Predict the values on both train and test
y_con_dec_tree_train_pred=con_dec_tree_model.predict(X_train_scaled)
y_con_dec_tree_test_pred=con_dec_tree_model.predict(X_test_scaled)
```

```python
#Report training accuracy and test accuracy
print('Training accuracy:',accuracy_score(y_clf_train,y_con_dec_tree_train_pred))
print('Test accuracy:',accuracy_score(y_clf_test,y_con_dec_tree_test_pred))
```

```python
# 3. Gini Decision Tree
gini_dec_tree_model=DecisionTreeClassifier(max_depth=5, criterion='gini', random_state=42)
gini_dec_tree_model.fit(X_train_scaled,y_clf_train)
```

```python
#Predict the values on both train and test
y_gini_dec_tree_train_pred=gini_dec_tree_model.predict(X_train_scaled)
y_gini_dec_tree_test_pred=gini_dec_tree_model.predict(X_test_scaled)
```

```python
#Report training accuracy and test accuracy
print('Training accuracy:',accuracy_score(y_clf_train,y_gini_dec_tree_train_pred))
print('Test accuracy:',accuracy_score(y_clf_test,y_gini_dec_tree_test_pred))
```

```python
print(classification_report(y_clf_test,y_gini_dec_tree_test_pred))
```

```python
#Entropy Decision Tree
en_dec_tree_model=DecisionTreeClassifier(max_depth=5, criterion='entropy',random_state=42)
en_dec_tree_model.fit(X_train_scaled,y_clf_train)
```

```python
#Predict the values on both train and test
y_en_dec_tree_train_pred=en_dec_tree_model.predict(X_train_scaled)
y_en_dec_tree_test_pred=en_dec_tree_model.predict(X_test_scaled)
```

```python
#Report training accuracy and test accuracy
print('Training accuracy:',accuracy_score(y_clf_train,y_en_dec_tree_train_pred))
print('Test accuracy:',accuracy_score(y_clf_test,y_en_dec_tree_test_pred))
```

```python
print(classification_report(y_clf_test,y_en_dec_tree_test_pred))
```

```python
# 4.Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
#train Random Forest model
ran_forest_model=RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
ran_forest_model.fit(X_train_scaled,y_clf_train)
```

```python
#Predict the values on both train and test
y_ran_forest_train_pred=ran_forest_model.predict(X_train_scaled)
y_ran_forest_test_pred=ran_forest_model.predict(X_test_scaled)
```

```python
#Report training accuracy and test accuracy
print('Training accuracy:',accuracy_score(y_clf_train,y_ran_forest_train_pred))
print('Test accuracy:',accuracy_score(y_clf_test,y_ran_forest_test_pred))
```

```python
#Roc_Auc score
print('Training roc_auc_score:',roc_auc_score(y_clf_train,y_ran_forest_train_pred))
print('Test roc_auc_score:',roc_auc_score(y_clf_test,y_ran_forest_test_pred))
```

```python
#model.feature_importances_

importances = ran_forest_model.feature_importances_

feature_names = X_train_scaled.columns

# Create a DataFrame and sort it in descending order
feature_imp_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

#  Extract the top 5 features
top_5_features = feature_imp_df.head(5)

# Display the scores as text
print("--- Top 5 Features by Importance ---")
print(top_5_features.to_string(index=False))
```

```python
# 4a. Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
```

```python
#Train the model
gra_model=GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gra_model.fit(X_train_scaled,y_clf_train)
```

```python
#Predict the values on both train and test
y_gra_train_pred=gra_model.predict(X_train_scaled)
y_gra_test_pred=gra_model.predict(X_test_scaled)
```

```python
#Report training accuracy and test accuracy
print('Training accuracy:',accuracy_score(y_clf_train,y_gra_train_pred))
print('Test accuracy:',accuracy_score(y_clf_test,y_gra_test_pred))
```

```python
#Roc_Auc score
print('Training roc_auc_score:',roc_auc_score(y_clf_train,y_gra_train_pred))
print('Test roc_auc_score:',roc_auc_score(y_clf_test,y_gra_test_pred))
```

```python
# Compute 5-Fold Cross-Validated
from sklearn.model_selection import  cross_val_score
cv_auc_scores = cross_val_score(gra_model, X, y_clf, cv=5, scoring='roc_auc')
mean_cv_auc = np.mean(cv_auc_scores)
mean_cv_auc
```

```python
 # 4b.Feature ablation study
 importance = ran_forest_model.feature_importances_
feature_name = X_train_scaled.columns

# Create a DataFrame and sort it in descending order
feature_imp_df = pd.DataFrame({
    'Feature': feature_name,
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

# Extract the lowest 5 features (the last 5 rows)
lowest_5_features = feature_imp_df.tail(5)

# Display the scores as text
print("--- Lowest 5 Features by Importance ---")
print(lowest_5_features.to_string(index=False))
```

```python
#Remove 5 columns from X_train_scaled and X_test_scaled
X_train_scaled_sec=X_train_scaled.drop(lowest_5_features['Feature'],axis=1)
X_test_scaled_sec=X_test_scaled.drop(lowest_5_features['Feature'],axis=1)
print(X_train_scaled_sec.shape)
print(X_test_scaled_sec.shape)
```

```python
#Train a second RandomForestClassifier
ran_forest_model_sec=RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
ran_forest_model_sec.fit(X_train_scaled_sec,y_clf_train)
```

```python
#Predict the values on both train and test
y_sec_ran_forest_train_pred=ran_forest_model_sec.predict(X_train_scaled_sec)
y_sec_ran_forest_test_pred=ran_forest_model_sec.predict(X_test_scaled_sec)
```

```python
#Report training accuracy and test accuracy
print('Training accuracy:',accuracy_score(y_clf_train,y_sec_ran_forest_train_pred))
print('Test accuracy:',accuracy_score(y_clf_test,y_sec_ran_forest_test_pred))
```

```python
#Roc_Auc score
print('Training roc_auc_score:',roc_auc_score(y_clf_train,y_sec_ran_forest_train_pred))
print('Test roc_auc_score:',roc_auc_score(y_clf_test,y_sec_ran_forest_test_pred))
```

```python
#Cross-validated comparison
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
models={
    "Decision Tree": dec_tree_model,
    "Decision Tree (controlled)": con_dec_tree_model,
    "Random Forest": ran_forest_model,
    "Gradient Boosting": gra_model
}
```

```python
#Evaluate each model using cross_val_score
results = []
for name, model in models.items():

    scores = cross_val_score(model, X, y_clf, cv=cv, scoring='roc_auc', n_jobs=-1)

    results.append({
        "Model": name,
        "Mean AUC": np.mean(scores),
        "Std Dev AUC": np.std(scores)
    })
```

```python
#Print the results
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# 6. Hyperparameter tuning with GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
#Build a pipeline
pipeline = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler(),
    RandomForestClassifier(random_state=42)
)
```

```python
#Define Paragrids
param_grid = {
    'randomforestclassifier__n_estimators': [50, 100, 200],
    'randomforestclassifier__max_depth': [5, 10, None],
    'randomforestclassifier__min_samples_leaf': [1, 5]
}
```

```python
#Combine pipeline and Paragrid
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=cv,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)
```

```python
#grid_search
grid_search.fit(X_train_scaled, y_clf_train)
```

```python
# Print the best hyperparameters and the corresponding best ROC-AUC score
print("Best Hyperparameters:")
for param, value in grid_search.best_params_.items():
    print(f"  {param}: {value}")

print(f"\nBest Cross-Validated ROC-AUC Score: {grid_search.best_score_:.4f}")
```

```python
#8. Serialize the best mode
import joblib

best_pipeline = grid_search.best_estimator_

joblib.dump(best_pipeline, 'best_model.pkl')
print("Best pipeline successfully saved to 'best_model.pkl'")
```

```python
#Short code
import joblib
import pandas as pd

# Reload the complete pipeline from disk
loaded_pipeline = joblib.load('best_model.pkl')

hand_crafted_data = pd.DataFrame([
    {col: X_train_scaled[col].median() for col in X_train_scaled.columns},
    {col: X_train_scaled[col].max() for col in X_train_scaled.columns}
])

predictions = loaded_pipeline.predict(hand_crafted_data)

print("Reloaded Model Verification Run:")
print(f"Predictions for hand-crafted rows: {predictions}")
```
